```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = '/content/song_dataset.csv'
df = pd.read_csv(file_path)
```

```python
# Display basic information about the dataset
print("Dataset Information:")
df.info()
```

```python
# Display the first few rows of the dataset
print("\nFirst few rows of the dataset:")
print(df.head())
```

#Most listened songs

This code identifies and prints the top 10 most listened songs using groupby and sort_values.

```python
most_listened_songs = df.groupby('song')['play_count'].sum().sort_values(ascending=False).head(10)

print(most_listened_songs)
```

#Plot Most listened songs.
The bar plot shows the song IDs on the x-axis and their total listen counts on the y-axis. The song with the highest bar is the most listened song.

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=most_listened_songs.index, y=most_listened_songs.values)
plt.title('Top 10 Most Listened Songs')
plt.xlabel('Song ID')
plt.ylabel('Total Listen Count')
plt.xticks(rotation=45)
plt.show()
```

# Most popular artists

This code identifies and prints the top 10 most popular artists sorted by total listen counts.

```python
most_popular_artists = df.groupby('artist_name')['play_count'].sum().sort_values(ascending=False).head(10)

print(most_popular_artists)
```

#Most Popular artists

The bar plot shows the artist names on the x-axis and their total listen counts on the y-axis. The artist with the highest bar is the most popular artist.

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=most_popular_artists.index, y=most_popular_artists.values)
plt.title('Top 10 Most Popular Artists')
plt.xlabel('Artist Name')
plt.ylabel('Total Listen Count')
plt.xticks(rotation=45)
plt.show()
```

# Distribution of song counts per user


This code calculates and prints the distribution of unique song counts per user using groupby, nunique, and histplot.

```python
user_song_count = df.groupby('user')['song'].nunique()

print(user_song_count)
```

#Distribution of songs count for users
The histogram shows the frequency of users listening to a certain number of unique songs.

```python
plt.figure(figsize=(10, 6))
sns.histplot(user_song_count, kde=True, bins=30)
plt.title('Distribution of Unique Song Counts per User')
plt.xlabel('Number of Unique Songs')
plt.ylabel('Frequency')
plt.show()
```
